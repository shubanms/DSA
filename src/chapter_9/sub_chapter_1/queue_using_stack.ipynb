# Implementation of Queue using Stack Data Structure

```python
from collections import deque

class Queue:
    def __init__(self):
        self.container_one = deque()
        self.container_two = deque()
        
    def insert(self, value):
        self.container_one.append(value)
        
    def remove(self):
        if(len(self.container_two) == 0):
            while(len(self.container_one) != 0):
                self.container_two.append(self.container_one.pop())
                        
        return self.container_two.pop()
    
    def peek(self):
        if(len(self.container_two) == 0):
            while(len(self.container_one) != 0):
                self.container_two.append(self.container_one.pop())
                        
        return self.container_two[-1]
    
    def size(self):
        return len(self.container_one)+len(self.container_two)
    
    def is_empty(self):
        if(len(self.container_two) == 0 and len(self.container_one) == 0):
            return True
        else:
            return False
```

```python
queue = Queue()
```

```python
queue.insert(1)
queue.insert(2)
queue.insert(3)
```

```python
queue.peek()
```
