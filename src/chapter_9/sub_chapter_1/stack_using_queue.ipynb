# Implementation of Stack using Queue Data Structure

```python
from collections import deque

class Stack:
    def __init__(self):
        self.buffer = deque()
        
    def push(self, value):
        self.buffer.append(value)
        while(self.buffer[0] != value):
            self.buffer.append(self.buffer.popleft())
            
    def pop(self):
        return self.buffer.popleft()
    
    def size(self):
        return len(self.buffer)
    
    def is_empty(self):
        return len(self.buffer) == 0
    
    def top(self):
        return self.buffer[0]
    
    def show(self):
        print(self.buffer)
```

```python
stack = Stack()
```

```python
stack.push(1)
stack.push(2)
stack.push(3)
```

```python
stack.show()
```

```python
stack.pop()
```

```python
stack.show()
```

```python
stack.top()
```

```python
stack.size()
```
