# Implementation of Queue using Arrays

```python
class Queue:
    def __init__(self):
        self.buffer = list()
        
    def insert(self, value):
        self.buffer.append(value)
        
    def remove(self):
        return self.buffer.pop(0)
    
    def peek(self):
        return self.buffer[0]
    
    def size(self):
        return len(self.buffer)
    
    def is_empty(self):
        return len(self.buffer) == 0
    
    def show(self):
        print(self.buffer)
```

```python
queue = Queue()
```

```python
queue.insert(1)
queue.insert(2)
queue.insert(3)
```

```python
queue.show()
```

```python
queue.remove()
```

```python
queue.show()
```

```python
queue.peek()
```
