# Implementation of Stack using Arrays

```python
class Stack:
    def __init__(self):
        self.container = list()
        
    def push(self, value):
        self.container.append(value)
        
    def pop(self):
        return self.container.pop()
    
    def size(self):
        return len(self.container)
    
    def is_empty(self):
        return len(self.container) == 0
    
    def top(self):
        return self.container[-1]
    
    def show(self):
        print(self.container)
```

```python
stack = Stack()
```

```python
stack.push(1)
stack.push(2)
stack.push(3)
```

```python
stack.top()
```

```python
stack.show()
```

```python
stack.pop()
```

```python
stack.show()
```
